Sad ću izvorne podatke prepisati tako da uzmem jednog reprezentanta grupe i sve članove grupe zapišem kao reprezentanta.

```python
reducirani = np.genfromtxt("300lip300walk_redalf.txt", dtype='str')
```

```python
def sestorke(s):
    a=0
    a=ak.index(s[5])*1+ak.index(s[4])*20+ak.index(s[3])*400+ak.index(s[2])*8000+ak.index(s[1])*160000+ak.index(s[0])*3200000
    return(a)

sest=[]
for i in range(64000000):
    sest.append(0)
    
f=open("300lip300walk_redalf.txt", "r")
lines=f.readlines()
for i in range(300):
    nl=lines[i+300].rstrip()
    k=len(nl)
    for j in range(k-6):
        a=nl[j:j+6]
        b=sestorke(a)
        sest[b]=sest[b]+1
f.close()
```

```python
def inverz_sestorke(br):
    slovo1=m.floor(br/3200000)
    res1=br%3200000
    slovo2=m.floor(res1/160000)
    res2=res1%160000
    slovo3=m.floor(res2/8000)
    res3=res2%8000
    slovo4=m.floor(res3/400)
    res4=res3%400
    slovo5=m.floor(res4/20)
    slovo6=res4%20
    
    
    return(ak[slovo1]+ak[slovo2]+ak[slovo3]+ak[slovo4]+ak[slovo5]+ak[slovo6])
```

```python
pod = {'Indeks':  range(64000000),
        'Frekvencija': sest}

df_sest = pd.DataFrame (pod, columns = ['Indeks','Frekvencija'])
df_sest = df_sest[df_sest['Frekvencija']>0]
df_sest['Ime'] = df_sest['Indeks'].apply(lambda x: inverz_sestorke(x))
df_sest.sort_values(by='Frekvencija', ascending=False, inplace=True)
```

Prvo ću provjeriti sve koji se sastoje samo od R ili I.

```python
f=open("300lip300walk.txt", "r")
lines=f.readlines()

rrrrrr=[]
for i in range(300):
    nl=lines[i+300].rstrip()
    rrrrrr.append(re.findall('[R|D|N|Q|E|K|I|L|M|V][R|D|N|Q|E|K|I|L|M|V][R|D|N|Q|E|K|I|L|M|V][R|D|N|Q|E|K|I|L|M|V][R|D|N|Q|E|K|I|L|M|V][R|D|N|Q|E|K|I|L|M|V]',nl))

f.close()

s=""
motiv=[]
for j in rrrrrr:
    for k in j:
        for i in k:
            s=s+i
        motiv.append(s)
        s=""
svi=c.Counter(motiv)
svi.most_common(20)
```
